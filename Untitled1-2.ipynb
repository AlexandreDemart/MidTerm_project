```python
import pandas as pd
import numpy as np
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from tqdm import tqdm
import matplotlib.pyplot as plt

import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset,Dataset
```

```python
data = pd.read_csv('../content/train.csv')
data.head()
```

```python
data.info() # there is nan in age
```

```python

avrAge=data["Age"].dropna().mean()
data["Age"] = data["Age"].fillna(avrAge) #remove nan in age with average value

# sex transform into 1 and 0
le=LabelEncoder()
le.fit(data["Sex"])
data["Sex"] = le.transform(data["Sex"])

data.head()
```

```python
# transform embarked harbour into 0, 1 and 2
le.fit(data["Embarked"])
data["Embarked"] = le.transform(data["Embarked"])
data.head()
```

```python
train, test = train_test_split(data, test_size=0.2, random_state=1, shuffle=True) #split data in 2
```

```python
train.head(),test.head()
```

```python
FEATURES = ["Pclass","Sex","Age","SibSp","Parch","Embarked"]
TARGET = "Survived"
```

```python
X_train=train[FEATURES]
Y_train = np.array(train[TARGET])
X_test=test[FEATURES]
Y_test = np.array(test[TARGET])
Y_test,X_train
```

```python
# I normalize my data
from sklearn.preprocessing import MinMaxScaler
scaller = MinMaxScaler()
```

```python
# normalization
X_train = scaller.fit_transform(X_train)
X_test = scaller.fit_transform(X_test)
X_train
```

```python
X_train = torch.from_numpy(X_train).float()
X_test = torch.from_numpy(X_test).float()
Y_train = torch.from_numpy(Y_train).long() 
Y_test = torch.from_numpy(Y_test).long()
# to have tensor
```

```python
model=Net() # model instance

criterion = nn.CrossEntropyLoss() # calculate loss function

optimizer = torch.optim.SGD(model.parameters(), lr=0.001) # optimizing weight
model
```

```python
#function to create dataSet

class dataSet(Dataset):
    
    def __init__(self,df):
        
  
        self.df = df
        
        self.train_X = np.array(self.df[FEATURES])
        self.train_Y = np.array(self.df[TARGET])
        
        self.train_X = scaller.fit_transform(self.train_X)
        
        self.train_X = torch.from_numpy(self.train_X).float()
        self.train_Y = torch.from_numpy(self.train_Y).long() 

    def __len__(self):
        
        return len(self.df)
    
    def __getitem__(self,idx):
        
        return {"X":self.train_X[idx],"Y":self.train_Y[idx]}
```

```python
train_dataset = dataSet(train)
test_dataset = dataSet(test)
```

```python
train_dataloader = DataLoader(train_dataset,batch_size=128,shuffle = True)
test_dataloader = DataLoader(test_dataset,batch_size=128,shuffle = False)
```

```python
for a in train_dataloader:
    print(a)
    break
```

```python
#function for train one epoch and have train losses 
def training(train_dataloader,model):

    total_loss = 0 # Initializing total loss
    
    model.train()

    for a in train_dataloader:

            train_x= a["X"]
            train_y = a["Y"]
            
            optimizer.zero_grad() # Set the gradient of optimizer to 0

            output = model(train_x) # prediction
            
            loss = criterion(output,train_y) # calculationg loss between predictions and answers

            loss.backward() # backward

            optimizer.step() # optimizing weight

            total_loss += loss.item() # integration of loss
            
    total_loss = total_loss/len(train_dataloader)
            
    return model,total_loss
        
```

```python
model,total_loss = training(train_dataloader,model)
```

```python
total_loss
```

```python
# function to calculate loss based on a model

def lossEvaluate(val_dataloader,model):
    

    total_loss = 0 # Initializing total loss

    model.eval()

    for a in val_dataloader:

        with torch.no_grad():


            val_x= a["X"]
            val_y = a["Y"]

            output = model(val_x) # prediction
                       
            loss = criterion(output,val_y) # calculationg loss between predictions and answers

            total_loss += loss.item() # integration of loss
            

    total_loss=total_loss/len(val_dataloader)
    
    
    return total_loss
```

```python
test_loss = lossEvaluate(test_dataloader,model)
test_loss
```

```python
#calculate accuracy based on our model
def calc_accuracy(x,y,model):
    preds = model(x)
    preds2 = torch.max(preds.data,1)[1]
    return accuracy_score(y,preds2)
```

```python
all_return = []
# train for 2000 epoch
for epoch in tqdm(range(2000)):
    
    model,train_loss = training(train_dataloader,model) #training of the model and train loss
    test_loss = lossEvaluate(test_dataloader,model) #loss of test
    train_score = calc_accuracy(X_train,Y_train,model) #accuracy of train 
    test_score = calc_accuracy(X_test,Y_test,model) #accuracy of test
    all_return.append([epoch,train_loss,test_loss,train_score,test_score])
```

```python
allData = pd.DataFrame(all_return)
allData.columns = ["epoch","train_loss","test_loss","train_score","test_score"]
allData
```

```python
# plot train and test loss
plt.plot(allData["epoch"],allData["train_loss"])
plt.plot(allData["epoch"],allData["test_loss"])
plt.legend(["train loss","test loss"])
plt.show()
```

```python
# plot train and test accuracy
plt.plot(allData["epoch"],allData["train_score"])
plt.plot(allData["epoch"],allData["test_score"])
plt.legend(["train accuracy","test accuracy"])
plt.show()
```
